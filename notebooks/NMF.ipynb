### Non-negative Matrix Factorization

```python
import pandas as pd
import numpy as np

from sklearn.decomposition import NMF
from sklearn.impute import SimpleImputer

import pickle

```

```python
base_path = '../Data/'
ratings = pd.read_csv(base_path + 'ratings.csv')
movies = pd.read_csv(base_path + 'movies.csv')
```

#### Rating Matrix (Parse)

```python
df = movies.merge(ratings,how='outer',on='movieId')
```

```python
df.head()
```

```python
df.columns
```

```python
dff = df[['movieId','userId','rating']]
```

```python
R_raw = dff.pivot(index='userId',columns='movieId',values='rating')
```

```python
R_raw.drop(R_raw.index[0],inplace=True)
```

```python
R_raw
```

```python
R_raw.to_csv('movies_user_ratings.csv')
```

```python
binary = pickle.dumps(R_raw)
open('movies_user_ratings.bin', 'wb').write(binary)
```

#### Factorization

```python
with open('movies_user_ratings.bin', 'rb') as f:
    df = pickle.load(f)
```

```python
df.head()
```

```python
imputer_zero = SimpleImputer(strategy="constant", fill_value=0)
```

```python
df_filled_zero = pd.DataFrame(imputer_zero.fit_transform(df), index =df.index, columns =df.columns)
```

```python
nmf= NMF(n_components=8,max_iter=500)
```

```python
binary = pickle.dumps(nmf)
open('nmf_zeros.bin', 'wb').write(binary)
```

```python
with open('nmf_zeros.bin', 'rb') as f:
    NMF = pickle.load(f)
```
